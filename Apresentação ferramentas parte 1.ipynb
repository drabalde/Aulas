## NumPy

Pacote para lidar com matrizes N-dimensionais, álgebra linear, números aleatórios dentre outros. É um "pacote fundamental" para computação científica em Python.

Alguns links úteis:

- http://www.numpy.org/

- https://docs.scipy.org/doc/numpy/user/quickstart.html

```
import numpy as np

a = np.arange(15)
a
```

```
print(a.shape)
a = a.reshape(-1,1)
print(a.shape)
a
```

```
b = np.arange(0,-15,-1).reshape(3,5)
b
```

```
a = a.reshape(3, 5)
c = a + b
c
```

```
n = np.arange(100,201)
n
```

```
n > 150
```

```
n[n > 150]
```
